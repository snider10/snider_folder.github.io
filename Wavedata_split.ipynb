```python
#ライブラリをインポート
import pandas as pd
```

```python
#解析データをdfとして取り込み
df = pd.read_excel("Wavedata.xlsx", sheet_name="wave", index_col=0)

#データの先頭5行を確認（データにミスが無いか確認）
df.head()
```

```python
x=0

while x < len(df.columns):

    df_2 = df.iloc[:,x:x+1]
    y = str(df.columns[x])
    df_2.to_csv(y + ".csv", index=False, header=False)
    x += 1
```
