# tsfreshで特徴量作成

```python
#ライブラリをインポート
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#解析データをdfとして取り込み、id=1の列を追加（id=1の行がtsfreshの解析対象になるので、全行にid=1を割り当てる）
df = pd.read_excel("入力データ.xlsx", sheet_name="波形データ", index_col=0)
nug = pd.read_excel("入力データ.xlsx", sheet_name="Nugget", index_col=0)

df['id'] = 1

#データの先頭5行を確認（データにミスが無いか確認）
df.head()
```

```python
#tsfreshからextract_features、MinimalFCParametersをインポート
from tsfresh import extract_features
from tsfresh.feature_extraction import extract_features, MinimalFCParameters

#欠損値補完用のimputeをインポート
from tsfresh.utilities.dataframe_functions import impute

#MinimalFCParameters関数はsettingsに入れておく
settings = MinimalFCParameters()
```

```python
#extract_featuresで特徴量を作成し、ts_featuresに格納
df_ts = extract_features(df, column_id='id')
df_ts
```

```python
#MinimalFCParametersで特徴量を作成し、Mini_featuresに格納
df_mini = extract_features(df, column_id="id", default_fc_parameters=settings)
df_mini
```

```python
#特徴量の中で計算不可だった値(NaN)を0で埋める
df_ts = impute(df_ts)
df_mini = impute(df_mini)
```

# 配列を整理

```python
df_ts = df_ts.transpose()
df_mini = df_mini.transpose()
```

```python
df_ts_al = pd.DataFrame(index=range(787), columns=[''])

target = '__'

x = 0

while x <  len(df_ts)/787:
    n = df_ts.index[x*787]
    idx = n.find(target)
    r = n[:idx]
    
    df_ts_al[r] = df_ts.iloc[x*787:(x+1)*787, 0].values
    x += 1


df_ts_al = df_ts_al.drop(df_ts_al.columns[0], axis=1)
```

```python
index_al = df_ts.index[0]

idx_2 = index_al.find(target)
r_2 = index_al[:idx_2+2:]


df_ts_id = df_ts.iloc[0:787,0]
df_ts_id = df_ts_id.rename(index=lambda s: s.replace(r_2, ""))

df_ts_al = df_ts_al.set_axis(df_ts_id.index, axis=0)


df_ts_al = df_ts_al.transpose()
```

```python
num = list(df_ts_al.index)
num = list(map(int, num)) 
df_ts_al['num'] = num

df_ts_al = df_ts_al.sort_values(by='num')
df_ts_al = df_ts_al.drop('num', axis = 1)

df_ts_al = df_ts_al.transpose()
```

```python
df_mini_al = pd.DataFrame(index=range(9), columns=[''])

target = '__'

x = 0

while x <  len(df_mini)/9:
    n = df_mini.index[x*9]
    idx = n.find(target)
    r = n[:idx]
    
    df_mini_al[r] = df_mini.iloc[x*9:(x+1)*9, 0].values
    x += 1


df_mini_al = df_mini_al.drop(df_mini_al.columns[0], axis=1)
```

```python
index_al = df_mini.index[0]

idx_2 = index_al.find(target)
r_2 = index_al[:idx_2+2:]


df_mini_id = df_mini.iloc[0:9,0]
df_mini_id = df_mini_id.rename(index=lambda s: s.replace(r_2, ""))

df_mini_al = df_mini_al.set_axis(df_mini_id.index, axis=0)


df_mini_al = df_mini_al.transpose()
```

```python
num = list(df_mini_al.index)
num = list(map(int, num)) 
df_mini_al['num'] = num

df_mini_al = df_mini_al.sort_values(by='num')
df_mini_al = df_mini_al.drop('num', axis = 1)

df_mini_al = df_mini_al.transpose()
```

## SelectKbest 上位の特徴量にフラグを立てる

```python
from sklearn.feature_selection import SelectKBest,f_regression

x = df_ts_al.transpose()
y = nug.transpose()
```

```python
selector = SelectKBest(score_func=f_regression, k=10) 
selector.fit(x, y)
mask = selector.get_support()
```

```python
df_ts_al.insert(0,"選別フラグ", mask)
```

```python
df_ts_al = df_ts_al.transpose()
df_ts_al.insert(0,"ナゲット径", y)
df_ts_al = df_ts_al.transpose()
```

# 結果を出力

```python
with pd.ExcelWriter('extracted_features.xlsx') as writer:
    df_ts_al.to_excel(writer, sheet_name='tsfresh特徴', encoding="cp932")
    df_mini_al.to_excel(writer, sheet_name='Minimal特徴', encoding="cp932")
```
